```python
import requests
from pprint import pprint
from lxml import html
import hashlib
from pymongo import MongoClient


client = MongoClient('localhost', 27017)
db = client['news_db']
t_news = db.news


news = []
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36', 'Accept': '*/*'}

###----------------------------------------------------------

main_link = 'https://yandex.ru/news/'
response = requests.get(main_link, headers=header)
dom = html.fromstring(response.text)

news_blocks = dom.xpath("//td[@class='stories-set__item']")
for block in news_blocks:
    news_data = {}
    news_data['_id'] = "https://yandex.ru/" + block.xpath(".//h2[@class='story__title']/a/@href")[0]
    news_data['source'] = "yandex.ru/news"
    news_data['name'] = block.xpath(".//h2[@class='story__title']/a/text()")[0]
    news_data['href'] = "https://yandex.ru/" + block.xpath(".//h2[@class='story__title']/a/@href")[0]
    news_data['time'] = block.xpath(".//div[@class='story__date']/text()")[0]
    news.append(news_data)


###----------------------------------------------------------

main_link = 'https://news.mail.ru/'
response = requests.get(main_link, headers=header)
dom = html.fromstring(response.text)

news_blocks = dom.xpath("//div[@class='daynews__item'] | //div[@class='daynews__item daynews__item_big']")
for block in news_blocks:
    news_data = {}
    news_data['_id'] = "https://news.mail.ru/" + block.xpath(".//a/@href")[0]
    news_data['source'] = "news.mail.ru"
    news_data['name'] = block.xpath(".//span/text()")[0]
    news_data['href'] = "https://news.mail.ru/" + block.xpath(".//a/@href")[0]
    news_data['time'] = ''
    news.append(news_data)

###----------------------------------------------------------


main_link = 'https://lenta.ru/'
response = requests.get(main_link, headers=header)
dom = html.fromstring(response.text)

news_blocks = dom.xpath("//div[@class='first-item'] | //div[@class='item']")
for block in news_blocks:
    news_data = {}
    news_data['_id'] = "https://lenta.ru/" + block.xpath(".//a/@href")[0]
    news_data['source'] = "lenta.ru"
    news_data['name'] = block.xpath(".//a/text()")[0]
    news_data['href'] = "https://lenta.ru/" + block.xpath(".//a/@href")[0]
    news_data['time'] = block.xpath(".//time/@title") + block.xpath(".//time/text()")
    news.append(news_data)

###----------------------------------------------------------

pprint(news)

for n in news:
    try:
        t_news.insert_one(n)
    except:
        continue


```
